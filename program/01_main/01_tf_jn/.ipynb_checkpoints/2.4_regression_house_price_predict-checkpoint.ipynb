```python
#导入包
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
#分成数据集和测试集
housing = fetch_california_housing()
# print(housing.data.shape)

x_train_all, x_test_raw, y_train_all, y_test = train_test_split(
    housing.data, housing.target, random_state = 7)
x_train_raw, x_valid_raw, y_train, y_valid = train_test_split(
    x_train_all, y_train_all, random_state = 11)


# x_train_all, x_test_raw= train_test_split(housing.data, random_state = 1)
# y_train_all, y_test = train_test_split(housing.target, random_state = 2)
# x_train_raw, x_valid_raw = train_test_split(x_train_all, random_state = 3)
# y_train, y_valid = train_test_split(y_train_all, random_state= 4)



#对数据进行标准化处理
sca = StandardScaler()
x_train = sca.fit_transform(x_train_raw)
x_valid = sca.transform(x_valid_raw)
x_test = sca.transform(x_test_raw)
```

## 对于input_shape,input_dim的理解
    神经网络需要输入单个样本的维度，而不需要总样本的数量（这是在fit时自动加进去的）。
    x_train.shape输出一个元组，第一个元素为数据的总量，后面的就是每个样本的维度，所以写为input_shape = x_train.shape[1:]
    input_shape需要等于一个可迭代对象。x_train,shape[1:]是进行切片，输出的是(8,),是一个元组；而x_train.shape[1]进行的是索引，输出的是
    一个整数。
    
    使用input_dim就需要直接指定维度，input_dim需要是一个整数，字符串或者字节类的对象。
    但还是用input_shape比较方便。

```python
model = keras.models.Sequential()

model.add(keras.layers.Dense(8,activation='relu', input_shape = x_train.shape[1:]))
model.add(keras.layers.Dense(1))

model.compile(loss = 'mean_squared_error', optimizer = 'adam')  
#因为是回归模型，所以判断模型好坏的标准就是均方差的大小，而不需要加上评价标准 metrics = ['accuracy']

history = model.fit(x_train, y_train,
          validation_data = (x_valid, y_valid),
          callbacks = [keras.callbacks.EarlyStopping(patience = 5, min_delta = 1e-3)],
          epochs = 20)

```

```python
def plot_learning_curves(history):
    pd.DataFrame(history.history).plot(figsize=(8, 5))
    plt.grid(True)
    plt.gca().set_ylim(0, 2)
    plt.show()
plot_learning_curves(history)

```

```python
model.evaluate(x_train, y_train, verbose = False)
```
