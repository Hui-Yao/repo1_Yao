# 7.索引，切片
    除了多维是要将所有索引都写在一个中括号内（用逗号隔开维数），其余语法完全与python相同
<font color = red>如何理解切片?</font>
    
    整数索引是相对于元素相对于左端的位置，复数索引是元素相对于右端的位置
    如果切片索引同号，那么你能推出所取出的元素数量；如果异号，则无法知道所取出的元素数量，因为相对坐标系都不样
<font color = red>索引与切片对于维度的影响？</font>

    每进行一次索引，都会使所得到的数据相对于原数组（列表）降低一个维度
    每进行一次切片，都会使所得到的数据保持其维度

```python
k = np.arange(5)
print(k)
print(k[1:-2])   #左边界是相对于左边的第二个数，右边界是相对与右边的第二个数。arange函数左闭右开
print(k[1:3])
print(k[-4:-2])
```

<font color = red>8.1 通过mask过滤取得所需的数据</font>

    1.mask过滤后得到由True和Flase组成的数组，也可以理解为1和0组成的数组，所以可以进行位运算
    位运算符：&（and）-全真为真，一假为假  |（or）：一真为真，全假为假  ～（not）：取反 ^（xor）：不同为真，相同为假
        结合使用多个mask时，也就是要符合两个过滤条件，所以要用到&，即为（mask1）&（mask2）
        
<font color = red>8.2 通过索引与切片取得所需的数据</font>

    clip的语法：clip（num1,num2）  将小于num1的全赋值为num1，将大于num2的全赋值为num2
    

```python
#通过mask
apple = np.array([1,2,3,4,5,6,7,8])
a1 = apple<7
a2 = apple>5

print(a1)
print('nonzero方法',np.nonzero(a1))
print(a2)
print('通过mask：',apple[a1])
print(apple[a1&a2])         #用&表示双重过滤


#通过切片与索引
apple = np.array([1,2,3,4,5,6,7,8])
apple[[1,2,5,6]] = 0
print('通过索引与切片：',apple)
```

# 12. 数组的广播原则
   <font color = red >如果两个数组的后缘维度(从末尾开始算起的维度)的轴长度相符或其中一方的长度为1，则认为它们是广播兼容的。广播会在缺失维度和(或)轴长度为1的维度上进行。如果缺少维度的话，则会在前面添加1来补全维度（其实也就是在外面加一层括号）。比如对于2\*3\*4的数组，2\*3\*1的可以与之计算，但是2\*3的不能，因为解释器会把2*3的当做 1\*2\*3的，那就是不匹配的。</font>
    
    通俗的理解是：设a = x*y*z维数组，数组b = p*q*r维数组，只要p，q,r相应为x,y,z或1，则数组a与b可以进行计算。
    

```python
import numpy as np
a = np.arange(24).reshape((2,3,4))

b = np.array([[[1]],[[1]]])    #b是2*1*1的数组
# print(b.shape)
c = np.array([[[1],[1],[1]],[[2],[2],[2]]])    #c是2*3*1
# print(c.shape)
print(a-b)
print(a-c)
```

# 13. 数组的轴
    如何理解数组的轴的方向？
        总结来讲：轴0的方向为最外层（最高层）数组元素的方向；轴1的方向为第二层数组元素的方向；以此类推。
        
        或者理解为：n维数组有n个轴（0到n-1）。一维数组只有一行，则轴0为行的方向；二维数组有一行和一列，则轴0代表行的方向，
        轴1代表列的方向。三维数组多了一个深度，此时轴0代表深度方向，1-行，2-列；即每增加一个新维度，轴0就代表该新维度方向，
        别的州的方向依次向后推。
        
    例子：
    一维数组：arr = [1,2,3] -> 最外层为第一个中括号，所以只有轴0；其内元素的方向即为1->2->3,所以轴0的方向也为1->2->3。
    二维数组：arr = [[1,2,3],
                   [4,5,6]]
    可视为：  arr = [x,
                    y], 其中x = [1,2,3], y = [4,5,6]
    所以最外层的方向即为x -> y,也即轴0的方向；轴1的方向为第二层数组的元素的方向，即为1->2->3
    
    通俗理解：三维数组可以理解为切一个立方体面包。先沿着轴0的方向，也即深度的方向把立方体面包切成片；
    再沿着轴1的方向，把片状面包切成条，最后沿着轴2的方向，将面包条切成丁。

## 16.1 np.random.rand(a,b,c,...)
    生成指定维度，且指定每个维度元素数量的数组。所有数据都属于[0,1)。
    括号里写了几个数字，生成的数组就是几维的。对应地，每一维上的数字规定了这一维的元素数量。
    每个元素的类型为np.float64  
    
## 16.2 np.random.randn()
    除了返回的数据服从标准正态分布，其余与rand（）相同。
    
## 16.3 np.random.random(size = None)
    随机返回浮点数组成的数组，元素的值属于[0,1）.
    random.random与前两个的区别在于前者的参数直接写就好了，而random要以元组的形式
  
## 16.4 np.random.randint(low, high=None, size=None, dtype=’l’)
    返回随机整数构成的数组，范围区间为[low,high），包含low，不包含high。low和high都要是整数
    参数：low为最小值，high为最大值，high没有填写时，默认生成随机数的范围是[0，low)；size为数组元素个数，这里数组的定义与rand（）相同；dtype为数据类型，默认的数据类型是np.int。
    
## 16.5 np.linspace(start,end,num)
    start代表起始的值，end表示结束的值，num表示在这个区间里生成数字的个数，生成的数组是等间隔生成的。
    start和end这两个数字可以是整数或者浮点数
    这个函数不属于random部分
    
## 16.6  np.random.normal(loc,scale,size)
    生成正态分布，而不是randn()的标准正态分布。randn()相当于normal(loc=0,scale=1)
    loc：float,此概率分布的均值（对应着整个分布的中心centre）
    scale：float,此概率分布的标准差（对应于分布的宽度，scale越大越矮胖，scale越小，越瘦高）
    size：int or tuple of ints,输出的shape，默认为None，只输出一个值

## 16.7 numpy.random.choice(a, size=None, replace=True, p=None)
    从给定的一维数组中生成随机数
    参数： a为一维数组类似数据或整数；size为数组维度；p为数组中的数据出现的概率，p的维度应与a相同。
    a为整数时，对应的一维数组为np.arange(a)

## 16.8 np.random.shuffle(array)
    参数是数组，对该数组随机改变顺序。如果是多维数组，则改变0轴的顺序。
    
## 16.9 np.random.seed()
    像一个语法糖，使随机可预测。
    放在同一个随机函数前面，seed里参数相同，那么不管运行多少次，生成的结果都是一样的。

   
   

```python
import numpy as np

x1 = np.random.rand(2,3)    #输出了一个三维数组，每一维度的元素数量对应了2,3
# print(type(x1[0][0]))

x2 = np.random.random((2,3))

x3 = np.random.randint(1,100,3,dtype='int64')
x4 = np.random.randint(1,10,(2,3))#生成一个二维随机数组，第一维2个元素，第二维3个元素






print('x1:\n',x1)
print('x2:\n',x2)
print('x3:\n',x3,'\n','x4:\n',x4)




```
